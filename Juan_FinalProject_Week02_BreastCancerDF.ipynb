```python
import pandas as pd
from sklearn.metrics import accuracy_score,confusion_matrix, roc_auc_score, roc_curve, f1_score
from sklearn.tree import DecisionTreeClassifier, export_graphviz 
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
import os
os.environ["PATH"] += os.pathsep + 'C:/Users/casti/Downloads/graphviz-2.38/release/bin'
import graphviz
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pprint
from mpl_toolkits.mplot3d import axes3d, Axes3D
```

```python
df = pd.read_csv('breast-cancer_csv.csv')
```

```python
df
```

```python
df.rename(columns = {'tumor-size':'tumorSize', 'inv-nodes':'invNodes','node-caps':'nodeCaps',
                     'deg-malig':'degMalig',
                     'breast-quad':'breastQuad'}, inplace = True)
df
```

```python
df['decade'] = df.age.str[0:2].astype(float)
df['tumorSizeMax'] = df.tumorSize.str[3:5]
df['tumorSizeMax'] = pd.to_numeric(df['tumorSizeMax'],errors='coerce')
df['invNodesMax'] = df.invNodes.str[2:3]
df['invNodesMax'] = pd.to_numeric(df['invNodesMax'],errors='coerce')
df = df[df.tumorSizeMax > 1]
df = df[df.invNodesMax > 1]
df
```

```python
df['menopause'].replace('premeno', 0, inplace=True)
df['menopause'].replace('ge40', 1, inplace=True)
df['menopause'].replace('lt40', 1, inplace=True)
df['irradiat'].replace('no', 1, inplace=True)
df['irradiat'].replace('yes', 0, inplace=True)
df['Class'].replace('no-recurrence-events', 1, inplace=True)
df['Class'].replace('recurrence-events', 0, inplace=True)
df = df.dropna()
df
```

```python
X = df[['decade','menopause', 'irradiat', 'degMalig','tumorSizeMax', 'invNodesMax']]
y = df['Class']
y.shape
```

## Logistic regression 

```python
split=0.19
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y, test_size=split)

m = LogisticRegression(solver='lbfgs')
m.fit(x_train, y_train)
X.shape
```

```python
m.score(X, y)
```

```python
m.score(x_train, y_train)
```

```python
m.score(x_test, y_test)
```

```python
y_pred = m.predict(x_test)
f1_score(y_test, y_pred)
X.shape
```

## Cross valitadion

```python
accuracy = cross_val_score(m, X, y, cv=5, scoring='accuracy')
print("cross-validation scores", accuracy)
X.shape
```

## Decision tree

```python
dm = DecisionTreeClassifier()
dm.fit(x_train,y_train)
y_pred_dm=dm.predict(x_test)
accuracy_score(y_test,y_pred_dm)
X.shape
```

```python
dm.score(X, y)
```

```python
dm.score(x_train, y_train)
```

```python
dm.score(x_test, y_test)
```

```python
tree = export_graphviz(dm, out_file=None,
                class_names=["cancer free", "cancer returned"],  
                feature_names=['decade','menopause', 'irradiat','degMalig', 'tumorSizeMax', 'invNodesMax'],
                impurity=False,
                filled=True)
open('breastcancer13.dot', 'w').write(tree)
```

```python
graph = graphviz.Source(tree)
graph.render('breastcancer13')
X.shape
```

```python
clf = RandomForestClassifier(n_estimators=7)
clf.get_params().keys()
```

```python
max_depth_list = [e for e in range(1,20)]
n_estimators_list = [i for i in range(1,100)]
param_grid = {
    'max_depth': max_depth_list,
    'n_estimators': n_estimators_list 
}
```

```python
cv = GridSearchCV(clf, param_grid, 
                  return_train_score=True, 
                  scoring='f1', 
                  cv=5, n_jobs=10)

cv.fit(x_train, y_train)
```
