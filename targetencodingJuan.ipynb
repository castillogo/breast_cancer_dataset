```python
df2 = pd.read_csv('pumps.csv',\
                 index_col=0)
```

```python
df3 = pd.read_csv('pumps_y.csv',\
                 index_col=0)
```

```python
df4 = df3.join(df2)
```

```python
ydummy = pd.get_dummies(df4['status_group'])
```

```python
ydummy
```

```python
df5 = pd.concat([df4, ydummy], axis=1)
```

```python
df5
```

```python
TM = df5.groupby('region').mean()
```

```python
TM
```

```python
df6 = df5
df6['region_functional'] = df6['region'].replace(TM['functional'])
df6['region_non functional'] = df6['region'].replace(TM['non functional'])
df6['region_functional needs repair'] = df6['region'].replace(TM['functional needs repair'])
```
